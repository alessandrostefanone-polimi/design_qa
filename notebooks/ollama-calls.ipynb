# Code

```python
from ollama import chat
```

```python
messages = []

def ollama_chat(user_input='hi!!!! :)', model='deepseek-r1'):
    
    message = {
        'role': 'user',
        'content': user_input,
    }
    
    messages.append(message)
    response = chat(model=model, messages=messages)
    
    print(response['message']['content'])
    
    return response
```

```python
query = "write me a python function that takes an input number and prints the two closest prime number to it: one smaller and one larger. Include a brief example at the end of the code."
response = ollama_chat(user_input=query, model='qwen3:4b')
```

```python
response['message']['content']
```

```python
from ollama import chat
```

```python
messages = []

def multimodal_chat(filepath, user_input='hi!!!! :)', model='deepseek-r1'):
    
    message = {
        'role': 'user',
        'content': user_input,
        'images': [filepath]
    }
    
    messages.append(message)
    response = chat(model=model, messages=messages)
    
    print(response['message']['content'])
    
    return response
```

```python
query = 'describe the image'
filepath = '../docs/images/1.png'
response = multimodal_chat(filepath=filepath, user_input=query, model='llava:latest')
```
